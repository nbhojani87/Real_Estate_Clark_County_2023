```python
# Create a new dataframe for all schools with the following columns
# Decided to forego including science info, since it's only avail for elementary schools and ELA and math will provide enough info
# ['Group', 'Organization Code', 'Mathematics - % Proficient','ELA - % Proficient']
# new dataframe is called ELA_Math_all
# no need to use a join key -- just need to pull in the appropriate columns
```

```python
# Join the two datasets, ELA_Math_Elementary and ELA_Math_High_School

# List of DataFrames
df = [ELA_Math_Elementary, ELA_Math_High_School]

# Concatenate DataFrames vertically
ELA_Math_all = pd.concat(df, axis=0, ignore_index=True)

# Print the combined DataFrame
ELA_Math_all.head()
```

```python
ELA_Math_all.info
```

```python
# Keep only the columns: ['Group', 'Organization Code', 'Mathematics - % Proficient','ELA - % Proficient']

ELA_Math_all = ELA_Math_all[['Group', 'Organization Code', 'Mathematics - % Proficient','ELA - % Proficient']]
```

```python
# Bring in zip code data.
# Read in Website_SchoolList

School_List_file_path = "/Users/mursil/workspace_neelam/Data_Projects_Personal/Real_Estate_Project/School Performance Information/Website_SchoolList.xlsx"

# # Load the excel file into a DataFrame
School_List = pd.read_excel(School_List_file_path)
```

```python
School_List.head(50)
```

```python
# Drop all rows where District != Clark County

# Define a condition
condition = School_List['District Name'] == 'Clark County'

# # Use the `drop()` function to drop rows that don't meet the condition
School_List = School_List.drop(School_List.index[~condition])

# # Print the resulting DataFrame
print(School_List)
```

```python
# join ELA_Math_all with School_List on ELA_Math_all ['Organization Code'] = School_List ['School Code']
# new dataframe = School_Perf_with_Address

School_Perf_with_Address = pd.merge(ELA_Math_all, School_List, 
                                    how='inner', 
                                    left_on='Organization Code', 
                                    right_on='School Code')
```
