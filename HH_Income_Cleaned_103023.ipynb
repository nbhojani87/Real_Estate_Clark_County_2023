```python
# read in HH income data by zip code, skip 1 row
# keep columns: zip code, HH mean income, HH median income
# extract numeric data from Geographic area name
```

```python
import pandas as pd
```

```python
# HH income data
HH_income_file_path = "/Users/mursil/workspace_neelam/Data_Projects_Personal/Real_Estate_Project/Household Income by Zip Code/ACSST5Y2021.S1901_2023-10-13T122844/ACSST5Y2021.S1901-Data.csv"

# # Read the csv file into DataFrame
HH_income = pd.read_csv(HH_income_file_path, skiprows=1)
```

```python
HH_income.head()
```

```python
HH_income = HH_income.drop(0, axis=0)
```

```python
HH_income.head()
```

```python
# Remove "ZCTA5 " from the 'Geographic Area Name' column and convert the result to numeric
HH_income['Geographic Area Name'] = HH_income['Geographic Area Name'].str.replace('ZCTA5 ', '').astype(int)
```

```python
HH_income.head()
```

```python
HH_income = HH_income.rename(columns={'Geographic Area Name': 'Zip_Code', 'Estimate!!Households!!Median income (dollars)' : 'Median_Income', 'Margin of Error!!Households!!Median income (dollars)' : 'Median_Income_Margin_of_Error',  'Estimate!!Households!!Mean income (dollars)' : 'Mean_Income', 'Margin of Error!!Households!!Mean income (dollars)' : 'Mean_Income_Margin_of_Error'})
```

```python
HH_income.head()
```

```python
HH_income = HH_income[['Zip_Code', 'Median_Income','Median_Income_Margin_of_Error', 'Mean_Income', 'Mean_Income_Margin_of_Error']]
```

```python
HH_income.head()
```

```python
HH_income.to_csv("HH_income.csv", index = False)
```
